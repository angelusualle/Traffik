```python
import numpy as np
import pandas as pd
```

```python
fatal = pd.read_csv('data/ingest/file_1_clean_age.csv')
states = pd.read_csv('data/ingest/us-state-names.tsv', sep = '\t')
population = pd.read_csv('populationsbystatebyyear.csv', index_col = False)
```

```python
fatal.head(5)
```

```python
states.head(5)
```

```python
data_working = pd.merge(fatal, states, how = 'left', left_on = 'STATE', right_on = 'id')
##print(data)
```

```python
data = data_working[['YEAR', 'name', 'FATALS']]
data = data.rename(columns = {'YEAR': 'date', 'name': 'STATE'})
##print(data)
```

```python
data['date'] = data['date'].astype(str) + '0101'
```

```python
pivot = data.pivot_table(values = 'FATALS', index = ['date'], columns = ['STATE'], aggfunc = np.sum)
```

```python
pivot['All States'] = pivot.sum(axis = 1)
```

```python
pivot
```

```python
pivot.to_csv('trend_fatalities.csv')
```

```python
pop = population.transpose()
new_header = pop.iloc[0]
state_population = pop[2:-1]
state_population.columns = new_header
state_population = state_population.rename(columns = {'State_Name': 'date'})
state_population
```

```python
per_capita = np.divide(pivot.drop(['All States'], axis = 1), state_population).astype(float)
per_capita['All States: Per Capita'] = per_capita.sum(axis = 1)
per_capita = per_capita.applymap('{:,.6f}'.format)
per_capita
```

```python
per_capita.to_csv('trend_percapita.csv')
```
